```python
missing_val = medstu_df.isnull().sum()
pd.DataFrame(missing_val)
```

```python
no_total_cells = np.product(medstu_df.shape)
print(f'The number of cells in the dataset is {no_total_cells}')
```

```python
no_missing_values = missing_val.sum()
print('The total number of missing values in the dataset is ', no_missing_values)
```

```python
missing_percentage = (no_missing_values/no_total_cells)*100
missing_percentage
```

```python
categorical_features = ['Gender','Blood Type','Diabetes','Smoking']
numerical_features = ['Student ID','Age','Height','Weight','BMI','Temperature','Heart Rate','Blood Pressure','Cholesterol']
# seperating categorical column and numerical columns for preprocessing
```

```python
categorical_df = pd.DataFrame(categorical_features)
categorical_df.info()
```

```python
numerical_df = pd.DataFrame(numerical_features)
numerical_df.info()
```

```python
for col in medstu_df.select_dtypes(include='object').columns: # Fill missing values in categorical columns with the mode
    mode_value = medstu_df[col].mode()[0]
    medstu_df[col] = medstu_df[col].fillna(mode_value)
```

```python
medstu_df.head()
```

```python
# Fill missing values in numeric columns with the mean
for col in medstu_df.select_dtypes(include='number').columns:
    mean_value = medstu_df[col].mean()
    medstu_df[col] = medstu_df[col].fillna(mean_value)
```

```python
medstu_df.head()
```

Hence the missing values were filled by ***mean and mode***

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Identify categorical columns
categorical_features = medstu_df.select_dtypes(include='object').columns
categorical_features
```

```python
for col in categorical_features:
    medstu_df[col] = le.fit_transform(medstu_df[col])
```
